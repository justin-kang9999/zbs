```python
# entropy
'''
p는 해당 데이터가 해당 클래스에 속할 확률
어떤 확률 분포로 일어나는 사건을 표현하는 데 필요한 정보의 양
이 값이 커질 수록 확률 분포의 불확실성이 커지며 결과 예측이 어려워짐
엔트로피는 이확률들의 합
'''
import numpy as np
p = np.arange(0.001, 1, 0.001)
plt.plot(p, -p*np.log2(p))
```

```python
# gini index
'''
불순도율
엔트로피의 계산량이 많음. 지니계수는 엔트로피와 비슷한 개념이면서 보다 계산량이 적음.
'''
```

```python
# sklearn을 이용한 decision tree
from sklearn.tree import DecisionTreeClassifier
# decision tree를 이용한 품종 분류
iris_tree = DecisionTreeClassifier()
# 학습시키기
iris_tree.fit(iris.data[:, 2:], iris.target)
```

```python
# accuracy 확인 분류 확률이 매우 높아 과적합 가능성이 존재
from sklearn.metrics import accuracy_score
y_pred_tr = iris_tree.predict(iris.data[:, 2:])
accuracy_score(iris.target, y_pred_tr)
```

```python
# mlxtend
# !pip install mlxtend
from mlxtend.plotting import plot_decision_regions
plt.figure(figsize=(14, 8))
plot_decision_regions(X=iris.data[:, 2:],
                      y = iris.target, clf = iris_tree, legend = 2)
# 경계면이 복잡해서 과적합 가능성이 존재함
```

```python
# 데이터 분리( 훈련용/ 검증용/ 평가용)
# 데이터를 훈련/ 테스트용으로 분리
from sklearn.model_selection import train_test_split
# 훈련용 데이터
features = iris.data[:, 2:] 
# 훈련용 데이터의 검증용 데이터
labels = iris.target
# 분리
X_train, X_test, y_train, y_test = train_test_split(features, labels,
                                                    test_size= 0.2,
                                                    random_state= 13)
```

```python
# 분리가 잘 되었는지 확인 - 클래스별로 동일비율이 아님
np.unique(y_test, return_counts=True)
```

```python
# 데이터를 훈련/ 테스트용으로 분리하는데 stratify 옵션을 이용
from sklearn.model_selection import train_test_split
# 훈련용 데이터
features = iris.data[:, 2:] 
# 훈련용 데이터의 검증용 데이터
labels = iris.target
# 분리
X_train, X_test, y_train, y_test = train_test_split(features, labels,
                                                    test_size= 0.2,
                                                    stratify= labels, 
                                                    random_state= 13)
```

```python
# 분리가 잘 되었는지 확인 - 클래스별로 동일비율로 분리가 됨
np.unique(y_test, return_counts=True)
```

```python
# 의사 결정 나무 모델 이용
# 일관성을 위해 random_state를 고정, 모델 단순화를 위해 max_depth 고정
iris_tree = DecisionTreeClassifier(max_depth= 2, random_state= 13) 
iris_tree.fit(X_train, y_train)
```

```python
# 정확도 확인
from sklearn.metrics import accuracy_score
y_pred_tr = iris_tree.predict(X_train)
accuracy_score(y_train, y_pred_tr)
```

```python
# 모델 확인
# !pip install graphviz
from graphviz import Source
from sklearn.tree import export_graphviz
Source(export_graphviz(iris_tree, feature_names=['length', 'width'],
                       class_names= iris.target_names,
                       rounded= True, filled= True))
```

```python
# 결정 경계 확인
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions

plt.figure(figsize=(12, 8))
plot_decision_regions(X= X_train,
                      y = y_train, clf = iris_tree, legend = 2)
```

```python
# 정확도 확인
y_pred_test = iris_tree.predict(X_test)
accuracy_score(y_test, y_pred= y_pred_test)
```

```python
# 잔기술 하나 - 전체 데이터에서 관찰해보기
scatter_highlight_kwargs = {'s': 150, 'label' : 'Test data', 'alpha' : 0.9}
scatter_kwargs = {'s' : 120, 'edgecolor' : None, 'alpha' : 0.7}
plt.figure(figsize=(12, 8))
plot_decision_regions(X= features, y = labels, 
                      X_highlight = X_test,
                      clf = iris_tree, legend = 2,
                      scatter_highlight_kwargs = scatter_highlight_kwargs,
                      scatter_kwargs = scatter_kwargs,
                      contourf_kwargs = {'alpha' : 0.2})
plt.show()
```
